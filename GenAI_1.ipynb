```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

np.random.seed(42)
samples_per_cluster = 300

cluster_a = np.random.normal(loc=2, scale=1, size=(samples_per_cluster, 2))
cluster_b = np.random.normal(loc=-2, scale=1, size=(samples_per_cluster, 2))
cluster_c = np.random.normal(loc=5, scale=1.5, size=(samples_per_cluster, 2))

all_data = np.concatenate((cluster_a, cluster_b, cluster_c), axis=0)

n_clusters = 3
model = GaussianMixture(n_components=n_clusters, covariance_type='full', random_state=42)
model.fit(all_data)
predicted_labels = model.predict(all_data)

plt.figure(figsize=(9, 6))
plt.scatter(all_data[:, 0], all_data[:, 1], c=predicted_labels, cmap='plasma', s=12, alpha=0.75)
plt.scatter(model.means_[:, 0], model.means_[:, 1], c='black', marker='P', s=120, label='Centers')
plt.title("Clustering with Gaussian Mixture Model", fontsize=14)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
